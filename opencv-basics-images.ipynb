```python
imgResize = imgFixToShowInPlt.copy()
imgResize = cv2.resize(imgResize, (300, 300))
plt.imshow(imgResize)
```

```python
imgFlip = imgFixToShowInPlt.copy()
imgFlip = cv2.flip(imgFlip, 1) # params (0, 1, -1)
plt.imshow(imgFlip)
```

```python
type(imgFlip)
```

```python
# recover color system before save, because opencv work with BGR

imgFixToSave = cv2.cvtColor(imgFlip, cv2.COLOR_BGR2RGB)
cv2.imwrite('data/0002-anonymous-flip.png', imgFixToSave)
```
