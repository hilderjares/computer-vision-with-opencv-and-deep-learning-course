```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_image = np.zeros(shape=(500, 500, 3), dtype=np.int16)
```

```python
blank_image.shape
```

```python
plt.imshow(blank_image)
```

```python
cv2.rectangle(blank_image, pt1=(400, 10), pt2=(420, 20), color=(0, 255, 255), thickness=5)
```

```python
plt.imshow(blank_image)
```

```python
cv2.circle(img=blank_image, center=(100, 100), radius=50, color=(255, 0, 0), thickness=8)
```

```python
plt.imshow(blank_image)
```

```python
cv2.circle(img=blank_image, center=(100, 400), radius=50, color=(255, 0, 0), thickness=-1)
plt.imshow(blank_image)
```

```python
cv2.line(blank_image, (0,0), (500,500), (255,255,0), 5)
plt.imshow(blank_image)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
image = blank_image.copy()

cv2.putText(
    image, 
    text='Hi OpenCV', 
    org=(20, 400), 
    fontFace=font, 
    fontScale=2, 
    color=(255, 255, 255),
    thickness=3,
    lineType=cv2.LINE_AA
)

plt.imshow(image)
```
